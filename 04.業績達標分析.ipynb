#### 【分析需求】

- **績效標竿識別**：從全體員工中篩選出總業績突破 **18,000,000 元** 的「優良員工」。
    
- **數據整合分析**：透過連結「員工基本資料」與「銷售交易紀錄」，將原本零散的訂單數據轉化為具備管理意義的人事績效指標。
    

#### 【結果解讀與商業價值】

- **核心獲利團隊**：清單中的員工為公司的「高產值人才」，其創造的業績是公司的主要營運來源。
    
- **決策建議**：
    
    1. **獎勵機制**：此名單可直接作為獎金發放、晉升評核或「年度最佳員工」的選拔依據。
        
    2. **資源分配**：評估是否給予績效優異員工更多的行銷資源或高潛力客戶名單，為公司創造更多營收。

```sql
SELECT e.Name as [員工姓名],
       SUM(s.Amount) as [總銷售業績]
  FROM HappyFruit.dbo.Employees as e
  JOIN HappyFruit.dbo.sales as s
    ON e.id=s.employee_id
 GROUP by e.Name
 HAVING SUM(s.Amount)>18000000
 ORDER by [總銷售業績] DESC;
```
