**【分析需求】**

- **基準數據清洗**<span style="color: var(--vscode-foreground);">：排除資料庫中「僅建檔但從無交易紀錄」的產品，篩選出市場真實流通的產品範疇，作為後續業績統計與精準報表的純淨基準。</span>
- **市場流通性識別**<span style="color: var(--vscode-foreground);">：</span><span style="color: var(--vscode-foreground);">透過EXISTS</span><span style="color: var(--vscode-foreground);">子查詢的過濾機制，精準識別出在市場上具備「實質交易紀錄」的品項清單。</span>

**【結果解讀與商業價值】**

- **歷史活躍品項定義**<span style="color: var(--vscode-foreground);">：名單內的產品代表其商業路徑已被市場驗證（曾產生過實質銷售），此清單可作為「具交易屬性商品」的基礎標籤，用於與停滯商品進行對照。</span>
- **營運調整建議**<span style="color: var(--vscode-foreground);">：</span>

1. <span style="color: var(--vscode-foreground);"><b>庫存管理</b></span><span style="color: var(--vscode-foreground);">：</span>對於此清單中之商品，建議進一步結合**時間刻度**進行分析，針對高頻交易商品建立安全庫存，預防缺貨帶來的營業損失。
2. <span style="color: var(--vscode-foreground);"><b>行銷策略</b></span><span style="color: var(--vscode-foreground);">：</span>此清單為後端分析的「基礎標籤」。可在此基準上進一步串接銷售金額與消費頻次，運用80/20法則篩選出真正驅動營收的明星產品。
3. <span style="color: var(--vscode-foreground);"><b>異常排除</b></span><span style="color: var(--vscode-foreground);">：</span>若行銷端正在主打的熱賣新品、或近期有安排曝光的品項「未出現在此清單中」，能立刻啟動反向追蹤機制，警示前端銷售系統可能存在資料串接遺漏。

```sql
SELECT p.Name as [產品]
  FROM HappyFruit.dbo.Products as p
 WHERE EXISTS (select 1
                     from HappyFruit.dbo.Sales as s
                    where p.id=s.product_id)
 ORDER by p.Name;
```
