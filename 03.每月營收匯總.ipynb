**【分析需求】**

- **非標準格式數據清洗**：針對 `Date` 欄位中含有「年、月、日」中文格式的非標準日期字串，進行數據格式轉化，使其具備可計算性。
    
- **年度業績關鍵指標 (KPI)**：鎖定 2019 年，將每日細碎的銷售資料彙整為「月度營收」，並設定 業績目標**5,000,000 元** 為優良業績門檻，藉此篩選出該年度的業績爆發月份。
    

【結果解讀與商業價值】

- **營收高峰識別**：清單中出現的月份，代表該月份的總銷售額突破了 500 萬大關，是公司該年度最主要的營收來源。
    
- **營運決策價值**：
    
    1. **季節性規律**：若篩選出的月份集中在特定時段，可證實產品具有明顯的季節性需求，應列為明年度重點備貨期。
        
    2. **資源配置**：強勢月份通常伴隨著高訂單量，未來應在此段期間增加物流與客服的人力支援。
        
    3. **異常排除**：若預期中的旺季（如過年、雙11）未出現在名單中，需檢討當時的行銷策略或庫存供應是否出現問題。

```sql
select YEAR(replace(replace(replace(date,'年','-'),'月','-'),'日','')) as [年份],
       MONTH(replace(replace(replace(date,'年','-'),'月','-'),'日','')) as [月份],
       SUM(Amount) as [總銷售額]
  from HappyFruit.dbo.Sales
 where YEAR(replace(replace(replace(date,'年','-'),'月','-'),'日',''))=2019
 group by YEAR(replace(replace(replace(date,'年','-'),'月','-'),'日','')),
          MONTH(replace(replace(replace(date,'年','-'),'月','-'),'日',''))
HAVING SUM(Amount)>5000000
 order by [月份];
 
```
