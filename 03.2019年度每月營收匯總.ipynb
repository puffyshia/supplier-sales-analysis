**【分析需求】**

- **非標準日期型態清洗與效能調整**：針對原始資料中DATE欄位以「年、月、日」中文格式儲存的非標準字串進行預處理。在實務大數據環境下，為避免重複進行字串替換導致全表掃描，應主動評估索引效能，或建議前端將其轉換為標準日期型態。
- **月度營收基準線過濾**：鎖定 2019 年度，將每日明細數據聚合為月度總額，並透過HAVING篩選出月營收突破5,000,000元的強勢月份，用以識別驅動大盤營收的核心區間。

【結果解讀與商業價值】

- **核心營收時段識別**：清單中呈現的月份為2019年營收貢獻的「前段班」。此數據能幫助管理階層客觀識別出公司的常態性旺季，作為跨年度營運規劃的基石。
- **營運決策價值**：

1. **供應鏈與物流配置：** 營收突破500萬的月份通常伴隨著訂單量暴增。營運端應以此清單為依據，提前規劃明年度相對應月份的庫存水位、物流彈性與客服人力支援，預防旺季塞單導致客戶體驗下降。
2. **行銷策略檢討：**由於HAVING屬於硬性過濾，若預期中的行銷旺季**未出現在名單中**，分析師應**主動調出未過濾的完整月營收**。確認是僅微幅落後門檻（調整促購力道），還是大幅落後（須檢討失誤原因）。

```sql
select YEAR(replace(replace(replace(date,'年','-'),'月','-'),'日','')) as [年份],
       MONTH(replace(replace(replace(date,'年','-'),'月','-'),'日','')) as [月份],
       SUM(Amount) as [總銷售額]
  from HappyFruit.dbo.Sales
 where YEAR(replace(replace(replace(date,'年','-'),'月','-'),'日',''))=2019
 group by YEAR(replace(replace(replace(date,'年','-'),'月','-'),'日','')),
          MONTH(replace(replace(replace(date,'年','-'),'月','-'),'日',''))
HAVING SUM(Amount)>5000000
 order by [月份];
 
```
