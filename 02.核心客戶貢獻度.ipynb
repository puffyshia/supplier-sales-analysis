#### 【分析需求】

- <span style="color: var(--vscode-foreground);"><b>客群價值總額化</b>：</span><span style="color: var(--vscode-foreground);">跳脫「單筆訂單」的破碎視角，以「客戶」為單位加總歷史消費金額，用以評估每位客戶的長期累積貢獻度。</span>
- **核心貢獻門檻建立**<span style="color: var(--vscode-foreground);">：</span><span style="color: var(--vscode-foreground);">運用HAVING</span><span style="color: var(--vscode-foreground);">搭配子查詢，計算出「已消費客群的平均消費額」作為業績基準線。精準篩選出高於平均值的核心貢獻者，排除低頻、低客單的邊緣長尾客群。</span>

#### 【結果解讀與商業價值】

- <span style="color: var(--vscode-foreground);"><b>高價值客群定義</b>：</span>清單中呈現的客戶，其累積消費力超越了已消費群體的平均水準。在電商營運中，這群人是驅動公司整體營收的基本盤，也是維繫品牌黏著度的核心受眾。
- **商業價值**<span style="color: var(--vscode-foreground);">：<br></span>

1. <span data-path-to-node="10,1,1,0,0" data-index-in-node="0" style="color: var(--vscode-foreground);"><span style="color: var(--vscode-foreground);"><b>行銷資源精準配置：</b></span></span><span style="color: var(--vscode-foreground);">本名單定義了高投報率的溝通對象。行銷部門應將有限的VIP專屬預算、尊榮客服資源或深度回饋機制集中投放於此，以最大化留存效益，防範核心資產流失。</span>
2. <span style="color: var(--vscode-foreground);"><b>升等策略：</b></span><span style="color: var(--vscode-foreground);">此平均值線可作為營運監控指標。後續行銷策略應引導「位於平均線稍微下方」的潛力客群跨越門檻、升等進入此高價值清單，進而拉升整體客單價與大盤營收。</span>

```sql
  SELECT c.Name as [客戶名稱],
        SUM(s.Amount) as [消費總金額]
    FROM HappyFruit.dbo.Clients as c
    JOIN HappyFruit.dbo.Sales as s
      ON c.id=s.client_id
  GROUP by c.Name
  HAVING SUM(s.Amount)>(select AVG(client_total)
                          FROM (SELECT SUM(Amount) as [client_total] --全體平均消費額
                                  FROM HappyFruit.dbo.Sales
                                  GROUP by client_id) as a)
    ORDER by [消費總金額] DESC;
```
