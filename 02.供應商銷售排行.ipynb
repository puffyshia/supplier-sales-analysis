#### 【分析需求】

- **建立業績基準線**<span style="color: var(--vscode-foreground);">：不以「單筆訂單」為單位，而是以「客戶」為單位進行彙整。</span>

- **篩選核心貢獻者**：找出總消費金額高於全體客戶平均值的 VIP 對象，排除掉零星購買或貢獻度較低的邊緣客戶。
    
- **資源配置調整**<span style="color: var(--vscode-foreground);">：將有限的行銷預算或客服資源優先投入在高貢獻客戶身上，實現 80/20 法則的精準管理。</span>
    

#### 【結果解讀與商業價值】

- **數據意義**<span style="color: var(--vscode-foreground);">：結果清單中的每一位客戶，其對公司的營收貢獻都屬於「前段班」。這群人的消費力是帶動整體業績向上提升的關鍵引擎。</span>
- **商業價值**<span style="color: var(--vscode-foreground);">：<br><ol><li><b data-path-to-node="10,1,1,0,0" data-index-in-node="0" style="color: var(--vscode-foreground);">資源集中化</b><span style="color: var(--vscode-foreground);">：這份名單就是行銷部門最應該「重點照顧」的對象（如：寄送節慶禮盒、提供 VIP 專屬折扣）。<br></span></li><li><b data-path-to-node="10,1,1,1,0" data-index-in-node="0" style="color: var(--vscode-foreground);">平均值拉升效應</b><span style="color: var(--vscode-foreground);">：如果我們能讓更多客戶從「平均線以下」跨越到這份名單內，公司整體營收將會顯著成長。</span></li></ol></span>

```sql
SELECT c.Name as [客戶名稱],
       SUM(s.Amount) as [消費總金額]
  FROM HappyFruit.dbo.Clients as c
  JOIN HappyFruit.dbo.Sales as s
    ON c.id=s.client_id
 GROUP by c.Name
 HAVING SUM(s.Amount)>(select AVG(client_total)
                         FROM (SELECT SUM(Amount) as [client_total]
                                 FROM HappyFruit.dbo.Sales
                                GROUP by client_id) as a)
  ORDER by [消費總金額] DESC;
```
