【分析需求】

- **精準報表呈現**：排除資料庫中「僅建檔但無交易」的產品（如過期測試資料、尚未上架之品項），避免統計時產生大量「銷售額為0」的雜訊，確保數據分析的純淨度。
    
- **市場流通性檢核**：識別出在市場上具有「實際交易紀錄」的品項。
    

【結果解讀與商業價值】

- **活躍資產定義**：名單內的產品視為公司的「活躍資產」，具有穩定的市場需求。
    
- **營運優化建議**：
    
    1. **庫存管理**：應針對此清單內的產品建立「充足庫存量」，預防缺貨造成損失。
        
    2. **行銷策略**：可針對此進一步分析哪些是公司的明星產品，加深銷售策略。
        
    3. **異常排除**：若名單中缺少了本應是熱賣的產品，則需回頭檢查前端銷售紀錄是否有串接漏洞。

```sql
SELECT p.Name
  FROM HappyFruit.dbo.Products as p
 WHERE EXISTS (select 1
                     from HappyFruit.dbo.Sales as s
                    where p.id=s.product_id)
 ORDER by p.Name;
```
